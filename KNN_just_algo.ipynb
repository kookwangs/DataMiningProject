# **KNN Predict from just 1 attribute**


```python

from collections import Counter
import math
import pandas as pd

def knn(data, query, k, distance_fn, choice_fn):
    neighbor_distances_and_indices = []
    
    # 3. For each example in the data
    for index, example in enumerate(data):
        # 3.1 Calculate the distance between the query example and the current
        # example from the data.
        
        print(example[:-1])
        distance = distance_fn(example[:-1], query)
        
        # 3.2 Add the distance and the index of the example to an ordered collection
        neighbor_distances_and_indices.append((distance, index))
    
    # 4. Sort the ordered collection of distances and indices from
    # smallest to largest (in ascending order) by the distances
    sorted_neighbor_distances_and_indices = sorted(neighbor_distances_and_indices)
    
    # 5. Pick the first K entries from the sorted collection
    k_nearest_distances_and_indices = sorted_neighbor_distances_and_indices[:k]
    
    # 6. Get the labels of the selected K entries
    k_nearest_labels = [data[i][-1] for distance, i in k_nearest_distances_and_indices]
    print("The k-nearest neighbor has these classes where k = ",k)
    print("\t\t",k_nearest_labels)
    print("Majority vote on them and get the result")
    # 7. If regression (choice_fn = mean), return the average of the K labels
    # 8. If classification (choice_fn = mode), return the mode of the K labels
    return k_nearest_distances_and_indices , choice_fn(k_nearest_labels)


def mean(labels):
    return sum(labels) / len(labels)

def mode(labels):
    return Counter(labels).most_common(1)[0][0]

def euclidean_distance(point1, point2):
    sum_squared_distance = 0
    for i in range(len(point1)):
        sum_squared_distance += math.pow(point1[i] - point2[i], 2)
    return math.sqrt(sum_squared_distance)

def main():
    data = pd.read_csv("preprocessed_data.csv")
    selected_column = "sit_and_bend_forward_cm"
    data1 = data[[selected_column,'class']]

    #data1 = data1.astype({'sepallength':'float','petallength':'float'})
    print("Preprocessed data: select only \"sit_and_bend_forward_cm\" and \"class\"")
    print(data1)

    data1 = data1.values.tolist()
    data1_predict = [9]
    print("Predict on the value ", data1_predict, " from ",selected_column)

    # Euclidean function

    print("Using Euclidean function")
    reg_k_nearest_neighbors, reg_prediction = knn(
        data1, data1_predict, k=7, distance_fn=euclidean_distance, choice_fn=mode
    )
    print("\t\toutput = ", reg_prediction)

if __name__ == '__main__':
    main()
```

# **KNN Predict from just 3 attribute**


```python

from collections import Counter
import math
import pandas as pd

def knn3(data, query, k, distance_fn, choice_fn):
    neighbor_distances_and_indices = []
    
    # 3. For each example in the data
    for index, example in enumerate(data):
        # 3.1 Calculate the distance between the query example and the current
        # example from the data.
        distance = distance_fn(example, query)
        
        # 3.2 Add the distance and the index of the example to an ordered collection
        neighbor_distances_and_indices.append((distance, index))
    
    # 4. Sort the ordered collection of distances and indices from
    # smallest to largest (in ascending order) by the distances
    sorted_neighbor_distances_and_indices = sorted(neighbor_distances_and_indices)
    
    # 5. Pick the first K entries from the sorted collection
    k_nearest_distances_and_indices = sorted_neighbor_distances_and_indices[:k]
    
    # 6. Get the labels of the selected K entries
    k_nearest_labels = [data[i][-1] for distance, i in k_nearest_distances_and_indices]
    print("The k-nearest neighbor has these classes where k = ",k)
    print("\t\t",k_nearest_labels)
    print("Majority vote on them and get the result")
    # 7. If regression (choice_fn = mean), return the average of the K labels
    # 8. If classification (choice_fn = mode), return the mode of the K labels
    return k_nearest_distances_and_indices , choice_fn(k_nearest_labels)

def mode(labels):
    return Counter(labels).most_common(1)[0][0]

def euclidean_distance_3(sample, target):
    sum_squared_distance = 0
    for i in range(3):
        sum_squared_distance += math.pow(sample[i] - target[i], 2)
    return math.sqrt(sum_squared_distance)

def main():
    data = pd.read_csv("preprocessed_data.csv")
    selected_column1 = "sit_and_bend_forward_cm"
    selected_column2 = "sit_ups_counts"
    selected_column3 = "broad_jump_cm"

    data1 = data[[selected_column1,selected_column2,selected_column3,'class']]
    print(data1)

    #data1 = data1.astype({'sepallength':'float','petallength':'float'})
    print("Preprocessed data: ")
    print(data1.dtypes)

    data1 = data1.values.tolist()
    data1_predict = [9,10,10]
    print("Predict on the value ", data1_predict)

    # Euclidean function

    print("Using Euclidean function")
    reg_k_nearest_neighbors, reg_prediction = knn3(
        data1, data1_predict, k=7, distance_fn=euclidean_distance_3, choice_fn=mode
    )
    print("\t\toutput = ", reg_prediction)

if __name__ == '__main__':
    main()
```
