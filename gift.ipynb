Line chart สำหรับค่า SystolicBP, DiastolicBP เทียบกับอายุ

Scatter plots ระหว่างน้าหนักและส่วนสูงของแต่ละเพศ และมีการระบุระดับสมรรถภาพทางกาย

```python
data4 = data[data["gender"] == 'M']
data4
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
colors = {'A':'red','B':'green','C':'blue','D':'pink'}
ax.scatter(x=data4['height_cm'],y=data4['weight_kg'],c=data4['class'].map(colors))
plt.show()
```
