```python
# Import data

import pandas as pd
data = pd.read_csv("preprocessed_data.csv")
data
```

## Data visualization

Bar chart เทียบว่าแต่ละ class มีค่าสุขภาพแต่ละตัวต่างกันอย่าง

```python
data2 = pd.DataFrame(data.groupby('class')[['height_cm','weight_kg','body_fat_%','diastolic','systolic','gripforce','sit_and_bend_forward_cm','sit_ups_counts','broad_jump_cm','bmi']].mean())
data2['class'] = data2.index
#data2['class'] = ['A','B','C','D']
data2
```

```python
data2.plot.bar(x='class',y=['height_cm','weight_kg','body_fat_%','diastolic','systolic','gripforce','sit_and_bend_forward_cm','sit_ups_counts','broad_jump_cm','bmi'],rot=0)
```

Line chart สำหรับค่า SystolicBP, DiastolicBP เทียบกับอายุ

```python
data3 = data.groupby('age')[['diastolic','systolic']].mean()
data3['age'] = data3.index

data3
```

```python
data3.plot.line(x='age',y=['systolic','diastolic'])
```

Scatter plots ระหว่างน้าหนักและส่วนสูงของแต่ละเพศ และมีการระบุระดับสมรรถภาพทางกาย

```python
data4 = data[data["gender"] == 'M']
data4
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
colors = {'A':'red','B':'green','C':'blue','D':'pink'}
ax.scatter(x=data4['height_cm'],y=data4['weight_kg'],c=data4['class'].map(colors))
plt.show()
```
