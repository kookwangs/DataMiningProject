```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("preprocessed_data.csv")
selected_column1 = ["age", "sit_ups_counts", "gender", "class"]
data1 = data[selected_column1]
data1_female = data1[data1["gender"]=="F"]
data1_male = data1[data1["gender"]=="M"]

fig, graph = plt.subplots()
colors = {'A':'green','B':'blue','C':'yellow','D':'red'}
graph.scatter(x=data1_female['age'],y=data1_female['sit_ups_counts'],c=data1_female['class'].map(colors))
plt.show()

fig, graph = plt.subplots()
graph.scatter(x=data1_male['age'],y=data1_male['sit_ups_counts'],c=data1_male['class'].map(colors))
plt.show()


```
