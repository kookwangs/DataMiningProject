## **Bar Chart of Gender by Class**

```python
# Group the dataframe by 'class' and 'gender', and count the number of entries in each group
grouped_df = data.groupby(['class', 'gender']).size().reset_index(name='count')

# Pivot the dataframe to create a new dataframe where the index is 'class', the columns are 'gender', and the values are the counts
pivot_df = grouped_df.pivot(index='class', columns='gender', values='count')

# Set the color palette to hls
sns.set_palette("Spectral")

# Create a stacked bar chart of the number of 'class' divided by 'gender' using matplotlib
pivot_df.plot(kind='bar', stacked=True)

# Set the title and labels of the plot
plt.title('Number of Class Divided by Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)

# Show the plot
plt.show()
```

```python
# Group the dataframe by 'class' and 'gender', and count the number of entries in each group
grouped_df = data.groupby(['class', 'gender']).size().reset_index(name='count')

# Set the color palette to hls
sns.set_palette("coolwarm")

# Create a bar chart of the number of 'class' divided by 'gender' using seaborn
sns.barplot(x='class', y='count', hue='gender', data=grouped_df)

# Set the title and labels of the plot
plt.title('Number of Class Divided by Gender')
plt.xlabel('Class')
plt.ylabel('Count')

# Show the plot
plt.show()
```

## **Bar Chart of Age by Class**

```python
# Define age ranges
age_ranges = [(1, 10), (11, 20), (21, 30), (31, 40), (41, 50), (51, 60), (61, 70), (71, 80)]

# Group the dataframe by 'class' and 'age', and count the number of entries in each group
grouped_df = data.groupby(['class', pd.cut(data['age'], [x[0] for x in age_ranges] + [age_ranges[-1][1]])]).size().reset_index(name='count')

# Pivot the dataframe to create a separate column for each age range
pivoted_df = grouped_df.pivot(index='class', columns='age', values='count')

# Set the color palette to hls
sns.set_palette("hls", 8)

# Create a stacked bar chart of the number of 'class' divided by range of 'age' using matplotlib
pivoted_df.plot(kind='bar', stacked=True)

# Set the title and labels of the plot
plt.title('Number of Class Divided by Age Range')
plt.xlabel('Class')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
# Define age ranges
age_ranges = [(1, 10), (11, 20), (21, 30), (31, 40), (41, 50), (51, 60), (61, 70), (71, 80)]

# Group the dataframe by 'age' and 'class', and count the number of entries in each group
grouped_df = data.groupby([pd.cut(data['age'], [x[0] for x in age_ranges] + [age_ranges[-1][1]]), 'class']).size().reset_index(name='count')

# Pivot the dataframe to create a separate column for each class
pivoted_df = grouped_df.pivot(index='age', columns='class', values='count')

# Set the color palette to hls
sns.set_palette("hls", 8)

# Create a stacked bar chart of the number of 'class' divided by range of 'age' using matplotlib
pivoted_df.plot(kind='bar', stacked=True)

# Set the title and labels of the plot
plt.title('Number of Class Divided by Age Range')
plt.xlabel('Age Range')
plt.ylabel('Count')

# Show the plot
plt.show()
```

Bar chart เทียบว่าแต่ละ class มีค่าสุขภาพแต่ละตัวต่างกันอย่าง

## **Scatter Plot between "Diastolic" and "Systolic" <br>with class label consider on each Gender**

```python
# Scatter plot (diastolic, systolic)

plt.rcParams["figure.figsize"] = [20,5]

fig, axs = plt.subplots(1, 2)

colors = {'A': 'lightgreen',
          'B': 'limegreen',
          'C': 'green',
          'D': 'darkgreen'}
markers = [plt.Line2D([0,0],[0,0],color=color, marker='o', linestyle='') for color in colors.values()]


axs[0].set_title("Male Diastolic and Systolic Blood Pressure Plot")
axs[0].set_xlabel("Diastolic blood pressure (mmHg)")
axs[0].set_ylabel("Systolic blood pressure (mmHg)")
axs[0].axhline(y=120, color='r', linestyle='-')
axs[0].axvline(x=80, color='r', linestyle='-')
axs[0].scatter(data[data['gender'] == 'M']['diastolic'], 
               data[data['gender'] == 'M']['systolic'],
               c=data[data['gender'] == 'M']['class'].map(colors), 
               label=data[data['gender'] == 'M']['class'])
axs[0].legend(markers, colors.keys(), loc="best")

axs[1].set_title("Female Diastolic and Systolic Blood Pressure Plot")
axs[1].set_xlabel("Diastolic blood pressure (mmHg)")
axs[1].set_ylabel("Systolic blood pressure (mmHg)")
axs[1].axhline(y=120, color='r', linestyle='-')
axs[1].axvline(x=80, color='r', linestyle='-')
axs[1].scatter(data[data['gender'] == 'F']['diastolic'], 
               data[data['gender'] == 'F']['systolic'],
               c=data[data['gender'] == 'F']['class'].map(colors), 
               label=data[data['gender'] == 'F']['class'])
axs[1].legend(markers, colors.keys(), loc="best")
```
